```python
import pandas as pd
import matplotlib.pyplot as plt
```

## eyes_open_look_fwd

```python
data = pd.read_csv("./eyes_open_look_fwd.csv")
data.drop("Unnamed: 0", axis = 1, inplace = True)
data['time']= pd.to_datetime(data['time']) 

framerate = ((data.time[data.shape[0]-1] - data.time[0])/data.shape[0]).total_seconds()
time_int= (data.time[data.shape[0]-1] - data.time[0]).total_seconds()
print('Number of datapoints:', data.shape[0], 
      '\nTime step:', framerate*1000, 'ms', 
      '\nFramerate:', 1/framerate, "/s" ,
      '\nTime interval:', time_int, 's')
```

### Blinking eyes left and right

```python
data[['bl_r', 'bl_l']].plot(figsize=(10,5))
plt.show()

data[['bl_r', 'bl_l']].hist(figsize=(10,5))
plt.show()
```

### Eyeballs left and right

```python
data[['ga_r', 'ga_l']].plot(figsize=(10,5))
plt.show()

data[['ga_r', 'ga_l']].hist(figsize=(10,5))
plt.show()
```
